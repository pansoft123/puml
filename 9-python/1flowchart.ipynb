```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, FancyArrow

def draw_flowchart():
    fig, ax = plt.subplots(figsize=(8, 6))
    
    def box(x, y, text):
        ax.add_patch(Rectangle((x, y), 2, 1, edgecolor='black', facecolor='lightblue'))
        ax.text(x + 1, y + 0.5, text, ha='center', va='center', fontsize=10)

    def arrow(x1, y1, x2, y2):
        ax.add_patch(FancyArrow(x1, y1, x2 - x1, y2 - y1, width=0.05, head_width=0.2, color='gray'))

    box(3, 6, 'Start')
    box(3, 4.5, 'Process Data')
    box(3, 3, 'Decision?')
    box(1, 1.5, 'Send Report')
    box(5, 1.5, 'End')

    arrow(4, 6, 4, 5.5)
    arrow(4, 4.5, 4, 4)
    arrow(4, 3, 2, 2.5)  
    arrow(4, 3, 6, 2.5) 

    ax.text(2.3, 2.6, 'No', fontsize=9)
    ax.text(5.3, 2.6, 'Yes', fontsize=9)

    ax.set_xlim(0, 8)
    ax.set_ylim(0, 8)
    ax.axis('off')
    plt.title("Flowchart with Matplotlib")
    plt.show()

draw_flowchart()
```
