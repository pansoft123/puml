```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Polygon

def make_rect(cx, cy, w, h):
    return Rectangle((cx - w/2, cy - h/2), w, h)

def make_diamond(cx, cy, s):
    return Polygon([(cx, cy - s/2), (cx + s/2, cy), (cx, cy + s/2), (cx - s/2, cy)])

fig, ax = plt.subplots()
ax.set_xlim(0, 4)
ax.set_ylim(5, 0)  
ax.axis('off')

# Start
ax.add_patch(make_rect(2, 0.5, 1.2, 0.6))
ax.text(2, 0.5, "Start", ha='center', va='center')

# Input
ax.add_patch(make_rect(2, 1.5, 1.2, 0.6))
ax.text(2, 1.5, "Input number", ha='center', va='center')

# Decision
ax.add_patch(make_diamond(2, 2.5, 1))
ax.text(2, 2.5, "Is even?", ha='center', va='center')

# Print Even
ax.add_patch(make_rect(1, 3.5, 1.2, 0.6))
ax.text(1, 3.5, "Print 'Even'", ha='center', va='center')

# Print Odd
ax.add_patch(make_rect(3, 3.5, 1.2, 0.6))
ax.text(3, 3.5, "Print 'Odd'", ha='center', va='center')

# End
ax.add_patch(make_rect(2, 4.5, 1.2, 0.6))
ax.text(2, 4.5, "End", ha='center', va='center')

# Arrows
ax.annotate('', xy=(2, 1.2), xytext=(2, 0.8), arrowprops=dict(arrowstyle='->'))
ax.annotate('', xy=(2, 2.0), xytext=(2, 1.8), arrowprops=dict(arrowstyle='->'))
ax.annotate('', xy=(1.6, 3.5), xytext=(1.5, 2.5), arrowprops=dict(arrowstyle='->'))
ax.annotate('', xy=(2.4, 3.5), xytext=(2.5, 2.5), arrowprops=dict(arrowstyle='->'))
ax.annotate('', xy=(2, 4.2), xytext=(1, 3.8), arrowprops=dict(arrowstyle='->'))
ax.annotate('', xy=(2, 4.2), xytext=(3, 3.8), arrowprops=dict(arrowstyle='->'))

# Labels for yes/no
ax.text(1.4, 3, "yes")
ax.text(2.6, 3, "no")

fig.set_size_inches(8, 10)
plt.show()
```
