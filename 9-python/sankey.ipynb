```python
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey

# Create a figure and axes
fig, ax = plt.subplots()

sankey = Sankey(ax=ax, scale=0.01, unit='units', format='%d')

sankey.add(
    flows=[100, -30, -70],  # 100 input, 30 work output, 70 heat loss output
    labels=['Input', 'Work', 'Heat Loss'],  # Labels for each flow
    orientations=[0, 0, 0],  # All flows are horizontal (left to right)
    patchlabel='Steam Engine'  # Label for the center of the diagram
)

# Finalize the Sankey diagram
sankey.finish()

# Set the title of the chart
plt.title('Steam Engine Efficiency')

# Save the plot to a file
plt.savefig('sankey_steam_engine.png')
```
